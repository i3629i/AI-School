## 2. 편향(Bias) 과 분산(Variance)

### 기댓값
    
기댓값은 나올 것이라고 예상하는 값을 기댓값이라고 한다.
E(X) = ∑P(x) * X

ex) 1등 상금이 1억원 복권 1장, 2등 상금이 100만원인 복권이 10장, 3등상금 1만원인 복권이 1000장 존재할때 1장을 샀을 때 기대할 수 있는 금액은??? (총 복권의 판매량은 백만장)

풀이 : X = [1억원, 100만원 , 1만원, 0원]
      P(X) = [1/1백만, 10/1백만, 1000/1백만, (여사건)]
      기댓값 ∑P(x) * X
      1억 * 1 / 백만 + 100만 * 10 / 백만 + 1만*1000 / 백만 + 0(여사건)
      100+10+10+0 = 120 <<기댓값
      
### 평균과 분산, 공분산

평균은 (전체값 / 구하고자 하는 갯수) 생략
편차 : 평균값과 데이터간에 얼마나 차이가 나는지, 평균값으로 부터의 차이

아래의 편차의 관점에서 보면 매달 얼마만큼 매출이 고객별로 흩어져 있는지 알 수 있다.
편차는 +방향과 - 방향 둘다 있어 값이 상쇄가 된다 이때 필요한 것이 분산이다.

분산 : 편차를 제곱한후 합계를 구하고, 이것을 다시 평균값으로 반든것(표준편차의 제곱) 이말은 분산이란 ∑(데이터-데이터평균)**2 의 평균

공분산 : 두 가지의 데이터에 대해서 상관관계를 가지는지 조사하기 위해 필요한 공분산이다. 


ε_x, ε_y는 각각 x데이터의 평균, y데이터의 평균

COV(X,Y) = 1/n * ∑(X - ε_x)*(Y - ε_y)

공분산을 계산할때는 단위에 대해서 신경을 쓸 필요 없다.(무게,신장)으로도 구할 수 있음.

공분산은 양의 값을 가질 때 양의 관계로 한쪽이 증가할때 다른한쪽도 증가
음의 값일때 한 쪽이 증가하면 한 쪽은 감소 한다는 관계를 가지고 있음.



```python
import pandas as pd
import numpy as np

data = {'1월' : [5000,10000,3000],
       '2월' : [5000,3000,7000],
       '3월' : [5000,1000,2000],
       '4월' : [5000,1000,8000],
       '5월' : [5000,15000,4000],
       '6월' : [5000,0,6000]}
mamazon = pd.DataFrame(data,index = ['백소연','이민준','이용진'])
#이런 데이터가 있을때
print(mamazon)

#편차는?
dev = mamazon.iloc[0].sum() / len(mamazon.iloc[0])
print(dev)
for i in range(0,3):
    for j in range(0,6):
        mamazon.iloc[i][j] -= dev
#평균값을 모두 빼준다
print(mamazon)

#분산은?
# 즉 분산은 편차의 제곱의 합의 평균을 구한것이다.
print( (mamazon.iloc[1]*mamazon.iloc[1]).sum() / len(mamazon.iloc[1]) )
```
