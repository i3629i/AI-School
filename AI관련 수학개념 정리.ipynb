## 1.확률변수와 확률분포
    이산확률변수 : 어떤변수 X를 P(X)의 확률로 나오게 할 수 있는. 
    ex) 주사위 2개를 던졌을때 확률변수 X = {2,3,4,5,6...12}
    
    이산확률분포 : 이산확률변수의 값에 따라 달라지는 확률들을 전체적으로 정리하여 나열한 것
    ex) P(X) = f(X)
    P(2) = 1/36 //두개의 합이 2가 나올 확률 f(x)
    P(7) = 6/36 //두개의 합이 7이 나올 확률
