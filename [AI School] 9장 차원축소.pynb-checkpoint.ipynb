## 차원의 저주 및 차원 축소

### 차원의 저주
    차원이 증가한다면 그것을 표현하기 위한 데이터 양이 기하급수 적으로 증가한다.
    고로 고차원의 데이터들의 의미를 표현하려면 차원 축소가 필요하다.
    일정 차원을 넘으면 분류한 성능이 0으로 수렴함.

    
### 차원 축소
    데이터의 의미를 제대로 표현할 특징을 추려내는 것.
    
    차원의 크기는 특징의 개수를 나타냄.
    특징이 너무 많아 학습이 어렵기 때문에 더 분류가 되는 좋은 특징만을 가지고 사용 한다는 의미
    
    Ex)인물 이미지에서의 모든 픽셀을 다 보지 않더라도 중요한 특징 픽셀만 잡아내서 분류가 가능함 이 특징을 잡는것이 중요!
